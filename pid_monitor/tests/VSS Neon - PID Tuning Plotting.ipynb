```python
import time
import psutil
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
plt.rcParams['animation.html'] = 'jshtml'
```

```python
fig = plt.figure(figsize=(10, 3))

ax_linear = fig.add_subplot(121)
ax_angular = fig.add_subplot(122)

fig.show()
```

```python
i = 0
x, y_l, y_a = [0 for _ in range(200)], [0 for _ in range(200)], [0 for _ in range(200)]

actual_message_size = -1

while True:
    
    logs_output = open('robot_output.txt').readlines()
    last_log = logs_output[-1]
    message_value = len(logs_output)
    
    if actual_message_size == message_value:
        continue
    else:
        actual_message_size = message_value
    
    try:
        linear, angular = last_log.split(';')
    except:
        continue
    
    x.append(i)
    y_a.append(float(angular))
    y_l.append(float(linear))
    
    x = x[-50:]
    y_a = y_a[-50:]
    y_l = y_l[-50:]
    
    ax_linear.clear()
    ax_angular.clear()
    
    ax_linear.plot(x, y_l, color='b')
    ax_angular.plot(x, y_a, color='r')
    
    fig.canvas.draw()
    
    time.sleep(0.01)
    i += 1
```
